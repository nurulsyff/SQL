# *JOIN*

## 1. Shows the ten cities with the largest population, sorted by the largest population.

```python
query = '''
select 
city.Name as city_name, 
country.Name as country_name, 
city.Population 

from city, country  

where city.CountryCode = country.Code 
order by city.population desc 
limit 10
;'''

pd.read_sql_query(query,engine)
```

## 2. The GNP, capital, and a city population of the Netherlands

```python
# implicite join
query='''
select 
country.Name as Country, 
country.GNP, 
city.Name as capital, 
city.population

from country, city
where country.code=city.countrycode
and country.name='Netherlands' 
and country.capital = city.id

;
'''
pd.read_sql_query(query,engine)
```

## 3. Ten countries with the highest percentage of Spanish speakers


```python
# implicite join
query='''
select c.name as Country, cl.percentage as Percentage
from country c, countrylanguage cl
where c.code =cl.countrycode
and cl.language='Spanish'
order by cl.percentage desc
limit 10
;
'''
pd.read_sql_query(query,engine)
```

```python
# inner join
query='''
select c.name as Country, cl.percentage as Percentage
from country c 
inner join countrylanguage cl
on c.code = cl.countrycode

where cl.language='Spanish'
order by cl.percentage desc
limit 10
;
'''
pd.read_sql_query(query,engine)
```

## 4. The GNP, the capital, the population of the capital and the official language spoken in Indonesia

```python
# using implicite join
query='''
select 
c.name as Country,
c.GNP,
city.name as Capital_City,
city.population as Population,
cl.language

from country c, city, countrylanguage cl

where city.countrycode=c.code 
and c.code=cl.countrycode

and c.name='Indonesia'
and cl.isofficial ='T'
and c.capital=city.ID
;
'''
pd.read_sql_query(query,engine)
```

```python
# using inner join
query='''
select 
country.name as Country,
country.GNP,
city.name as Capital_City,
city.population as Population,
countrylanguage.language

from country 
join city 
on country.code=city.countrycode
join countrylanguage 
on country.code=countrylanguage.countrycode

where country.name='Indonesia'
and countrylanguage.isofficial ='T'
and country.capital=city.ID
;
'''
pd.read_sql_query(query,engine)
```

# *SUB QUERY*

## 5. The continent with more countries than North America


```python
query = '''
select continent, count(name) as Total_Country 
from country
group by continent
having count(name) > 
                    (select count(name) 
                    from country 
                    where continent = 'North America' )
;'''
pd.read_sql_query(query,engine)
```

```python
# number of countries continent in North America
query = '''
select count(name) as Countries_NorthAmerica
from country 
where continent = 'North America'
;'''
pd.read_sql_query(query,engine)
```

## 6. Countries in Asia with a GNP above the average GNP of countries on the European continent, sorted from largest.


```python
query='''
select name as Country, GNP 
from country

where GNP > 
            (select avg(GNP) 
            from country 
            where continent='Europe')

and continent = 'Asia'
order by GNP desc
;
'''
pd.read_sql_query(query,engine)
```

```python
# Europe's average GNP (1st method)
query='''
select avg(GNP)  as Average_GNP
from country
where continent='Europe'
;'''
pd.read_sql_query(query, engine)
```

```python
# Europe's average GNP (2nd method)
query='''
select continent, avg(GNP) from country
group by continent
having continent='Europe'
;'''
pd.read_sql_query(query, engine)
```

```python
# Countries with GNP above the average Europe's GNP
query='''
select name, GNP from country

where GNP > (select avg(GNP) from country
group by continent
having continent='Europe')
and continent='Asia'
order by GNP desc
;'''
pd.read_sql_query(query, engine)
```

## 7. Continents ending with the letter 'a' and the number of regions have more than the number of regions of the Asian continent.

```python
# total region in Asia
query = '''
select count(distinct(region)) as total_region
from country 
group by continent 
having continent='Asia'
;'''
pd.read_sql_query(query,engine)

```

```python
# total region by Continents
query='''
select 
count(distinct(region)) as total_region, 
continent 
from country

where continent like '%a'
group by continent 

having count(distinct(region)) > 
                                (select count(distinct(region)) 
                                from country 
                                group by continent 
                                having continent='Asia')

;'''
pd.read_sql_query(query, engine)
```

## 8. Five countries with the largest GNP have English as the official language, and the language percentage is above 50%.

```python
query='''
select name as Country, GNP 
from country 

where Code in 
    (select CountryCode from
    countrylanguage 
    where Language = 'English' and IsOfficial = 'T' and Percentage > 50) 

order by GNP desc 
limit 5
;'''
pd.read_sql_query(query, engine)
```

## 9. The list of Asian countries has a life expectancy of more than the average life expectancy of European countries.

```python
query='''
select 
Name as Name, 
continent as Continent, 
LifeExpectancy as Life_Expectancy, 
GNP 

from country 

where Continent='Asia' and LifeExpectancy > 
    (select avg(LifeExpectancy) 
    from country
    where continent='Europe') 

order by LifeExpectancy desc;

;'''
pd.read_sql_query(query, engine)
```

## 10. The list of ASEAN countries and their population, Gross National Product (GNP), capital name, and capital's population is sorted alphabetically by country name.

```python
query='''
SELECT 
a.name as Country, 
a.population as Country_Population, 
a.GNP, b.name as CapitalCity,
b.population as CapitalCity_Population

FROM COUNTRY a 
JOIN city b ON a.Capital = b.ID

WHERE a.region = 'SouthEast Asia' 
order by 1;

;'''
pd.read_sql_query(query, engine)
```
